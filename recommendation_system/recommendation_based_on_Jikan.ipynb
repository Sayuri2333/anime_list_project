```python
import numpy as np
def get_recommend(animelist):
    animelist = [int(x) for x in animelist]
    s = requests.session()
    url_pattern = "https://api.jikan.moe/v3/anime/{}/recommendations"
    counts = {}
    for i in animelist:
        print(i)
        url = url_pattern.format(i)
        html = s.get(url).json()
        for id_count in html['recommendations']:
            id = id_count['mal_id']
            count = id_count['recommendation_count']
            counts[id] = counts.setdefault(id, 0) + count
    counts_list=[]
    for key, value in counts.items():
        counts_list.append((key, value))
    data = pd.DataFrame(counts_list)
    data.columns = ['id', 'cnt']
    data = data.sort_values(by='cnt' , ascending=False).head(10)[['id']]
    data = set([x[0] for x in data.values]) - set(animelist)
    return data
```

```python
data = get_recommend(['1', '205', '6'])
data
```
